## LR  Using MLE

```python
# Step 1: imports
import numpy as np
from scipy.optimize import minimize
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Step 2: Input Data
x = np.array([1, 2, 3])
y = np.array([2, 3, 5])
```

```python
# Step 3: Negative Log Likelihood function
def neg_log_likelihood(params):
    w, b = params
    sigma2 = 1   # variance
    y_pred = w * x + b
    nll = 0.5 * np.sum((y - y_pred) ** 2 / sigma2)
    return nll
```

```python
# Step 4: Initial guess
initial_guess = [0, 0]
```

```python
# Step 5: Minimize NLL
result = minimize(neg_log_likelihood, initial_guess)
w_mle, b_mle = result.x

print(f"Slope is: {w_mle}")
print(f"Intercept is: {b_mle}")
```

```python
# Step 6: Prediction
y_pred = w_mle * x + b_mle

print("\nPrediction for Training Data")
for xi, yi, ypi in zip(x.flatten(), y, y_pred):
    print(f"\nx={xi}\ny={yi}\nPredicted y={ypi}")
```

```python
# Step 7: Error Calculate
mse=mean_squared_error(y,y_pred)
r2=r2_score(y,y_pred)
print(f"Mean Squared Error is:{mse}")
print(f"R2_Square is:{r2}")
```

## GD


```python
# Step 1: imports
import numpy as np
from scipy.optimize import minimize
```

```python
# Step 2: Input Data
x = np.array([1, 2, 3])
y = np.array([2, 3, 5])
```

```python
# Step 3: Initialize parameters
w,b=0,0
alpha =0.001
m=10000
n=len(x)
```

```python
# Step 4: Gradient Descent loop
for i in range(m):
    y_pred = w * x.flatten() + b
    dw = (-2/n) * np.sum(x.flatten() * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)
    w = w - alpha * dw
    b = b - alpha * db

print(f"Slope is:{w}")
print(f"Bias is:{b}")
```

```python
# Step 6: Prediction
y_pred = w * x + b

print("\nPrediction for Training Data")
for xi, yi, ypi in zip(x, y, y_pred):
    print(f"x={xi}, y={yi}, Predicted y={ypi}")
```

```python
# Step 7: Error Calculate
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"\nMean Squared Error is: {mse}")
print(f"R2_Square is: {r2}")
```

```python
# Step 1: imports
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
```

```python
# Step 2: Input Data
x = np.array([1, 2, 3])
y = np.array([2, 3, 5])
n=len(x)
```

```python
# Step 3: Calculate Loss Function
def loss(w1):
    w0=np.mean(y)-w1*np.mean(x)
    y_pred = w1*x + w0
    return np.sum((y-y_pred)**2)
```

```python
# Step 4: Calculate gradient of 3 wrt w1
def gradient(w1):
    w0=np.mean(y)-w1*np.mean(x)
    y_pred=w1*x+w0
    return -2/n*np.sum(x*(y-y_pred))
```

```python
# Step 5: Gradient Descent
lr=0.1
w1=4
iter=15
w1_values=[]
loss_values=[]

for i in range(iter):
    w1_values.append(w1)
    loss_values.append(loss(w1))
    grad=gradient(w1)
    w1 = w1 - lr * grad
```

```python
# Step 6: Plotting
w_space=np.linspace(-2,4,200)
loss_space=[loss(w) for w in w_space]
plt.figure(figsize=(8,6))
plt.plot(w_space, loss_space, label="Loss Curve")
plt.scatter(w1_values, loss_values,color="Red")
plt.xlabel("Slope (w1)")
plt.ylabel("Loss")
plt.title("Gradient Descent Optimization (Single Parameter)")
plt.show()
```
